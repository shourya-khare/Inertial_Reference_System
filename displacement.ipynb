```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "accelerometer-disp.csv"
df = pd.read_csv(data)
dataset = df.to_numpy()
t, ax, ay, az, amag = dataset.T
print(t[10])
```

```python
plt.xlabel("t [s]")
plt.ylabel("a [m/s^2]")
plt.plot(t, ax, label="ax")
plt.plot(t, ay, label="ay")
plt.plot(t, az, label="az")
plt.grid()
plt.legend()
plt.show()
```

```python
dt = t[1]-t[0]
vx = [0]
vy = [0]
vz = [0]

for i in np.arange(len(t)-1):
    vx = vx + [vx[-1] + ax[i]*(t[i+1]-t[i])]
    vy = vy + [vy[-1] + ay[i]*(t[i+1]-t[i])]
    vz = vz + [vz[-1] + (az[i]-9.88)*(t[i+1]-t[i])]

plt.plot(t, vx, label="vx")
plt.plot(t, vy, label="vy")
plt.plot(t, vz, label="vz")

plt.xlabel("t [s]")
plt.ylabel("v [m/s]")
plt.grid()
plt.legend()
plt.show()
```

```python
x = [0]
y = [0]
z = [0]

for i in np.arange(len(t)-1):
    x = x + [x[-1] + vx[i]*(t[i+1]-t[i])]
    y = y + [y[-1] + vy[i]*(t[i+1]-t[i])]
    z = z + [z[-1] + (vz[i])*(t[i+1]-t[i])]

plt.plot(t, vx, label="x")
plt.plot(t, vy, label="y")
plt.plot(t, vz, label="z")

plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.grid()
plt.legend()
plt.show()
```
